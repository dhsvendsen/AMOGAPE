### L8 results 100 updates all methods

```python
L8methods = ['add_D', 'add_G', 'add_D_add_G', 'add_D_prod_G' ,'prod_D_prod_G','prod_D_add_G','gauss','lhs']
L8testers = ['add_D_test', 'add_G_test', 'add_D_add_G_test', 'add_D_prod_G_test','prod_D_prod_G_test','prod_D_add_G_test','gauss_test','lhs_test']
L8trackers = ['add_D_tracker', 'add_G_tracker', 'add_D_add_G_tracker', 'add_D_prod_G_tracker' ,'prod_D_prod_G_tracker','prod_D_add_G_tracker','gauss_tracker','lhstracker']
```

```python
with open('resultsL8.pkl','rb') as handle:
    resultsL8 = pickle.load(handle)
```

```python
runs = 15
measurements = (100)/5 +1 #11
endpoints = 130
resdict = {method:None for method in L8methods}
for method in L8methods:
    resultmat = np.zeros([runs,measurements])
    for run in range(runs):
        resultmat[run,:] = resultsL8[run][method+'_test']
    resdict[method] = resultmat.mean(axis=0)
```

```python
plt.figure(figsize=(9,5))
  
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['add_D'], '-^', label = '$\Sigma D$', ms=5)
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['prod_D'], '-o', label = '$\Pi D$', ms=3, color='C2')
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['add_D_add_G'], '-*', label = '$\Sigma D$ x $\Sigma G$', ms=6, color='C3')
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['add_D_prod_G'], '-v', label = '$\Sigma D$ x $\Pi G$', ms=5, color='C4')
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['prod_D_add_G'], '-<', label = '$\Pi D$ x  $\Sigma G$', ms=5, color='C5')
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['prod_D_prod_G'], '->', label = '$\Pi D$ x $\Pi G$', ms=5, color='C6')
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['gauss'],'--' ,label = 'Random', color='C1',ms=3)
        
plt.legend()
plt.ylabel('RMSE',fontsize=16)
plt.xlabel('Number of nodes ($m_t$)',fontsize=16)
plt.yticks([0.1,1],fontsize=14)
plt.xticks(np.linspace(30,endpoints,11),fontsize=14)
plt.legend(fontsize=12)
```

### 3D results

```python
L8methods = ['add_D' ,'prod_D_prod_G','gauss']
L8testers = ['add_D_test', 'add_G_test', 'add_D_add_G_test', 'add_D_prod_G_test','prod_D_prod_G_test','prod_D_add_G_test','gauss_test','lhs_test']
L8trackers = ['add_D_tracker', 'add_G_tracker', 'add_D_add_G_tracker', 'add_D_prod_G_tracker' ,'prod_D_prod_G_tracker','prod_D_add_G_tracker','gauss_tracker','lhstracker']
```

```python
with open('resultsL8_freenoise_125_3D_1run.pkl', 'rb') as handle:
    resultsL8_freenoise125_3d_v1 = pickle.load(handle) 
    
with open('resultsL8_freenoise_125_3D_1run_2.pkl', 'rb') as handle:
    resultsL8_freenoise125_3d_v2 = pickle.load(handle) 
    
with open('resultsL8_freenoise_125_3D_5run.pkl', 'rb') as handle:
    resultsL8_freenoise125_3d_v3 = pickle.load(handle) 
```

```python
resultsL8_freenoise125_3d = resultsL8_freenoise125_3d_v3
resultsL8_freenoise125_3d[5] = resultsL8_freenoise125_3d_v1[0]
resultsL8_freenoise125_3d[6] = resultsL8_freenoise125_3d_v2[0]
```

```python
runs = 7
measurements = (125)/5 +1 #11
endpoints = 156
resdict = {method:None for method in L8methods}
for method in L8methods:
    resultmat = np.zeros([runs,measurements])
    for run in range(runs):
        resultmat[run,:] = resultsL8_freenoise125_3d[run][method+'_test']
    resdict[method] = resultmat.mean(axis=0)
```

```python
plt.figure(figsize=(7.5,4))

plt.semilogy(np.linspace(30,endpoints,measurements), resdict['add_D'], '-^', label = '$\Sigma$ D', ms=6)
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['gauss'],'--' ,label = 'Random')
plt.semilogy(np.linspace(30,endpoints,measurements), resdict['prod_D_prod_G'], '->', label = '$\Pi$ D x $\Pi$ G', ms=6, color='C6')

plt.legend()
plt.ylabel('RMSE',fontsize=16)
plt.xlabel('Number of nodes ($m_t$)',fontsize=16)
#plt.yticks([6e-1, 1, 2,3,4],fontsize=14)
plt.xticks(np.linspace(30,endpoints,11),fontsize=14)
plt.legend(fontsize=12)

plt.savefig('3dresults.pdf',format='pdf',bbox_inches='tight')
```
